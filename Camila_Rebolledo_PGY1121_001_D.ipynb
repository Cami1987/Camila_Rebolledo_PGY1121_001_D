**Evaluación Parcial 3: Arreglos y Funciones**


```python
autos= []

def grabar():
    tipo=input("Ingrese el tipo de vehículo: ")
    patente=input("Ingrese la patente del vehículo: ")
    marca=input("Ingrese la marca del vehículo: ")
    while len(marca) < 2 or len(marca) > 15:
        marca = input("La marca debe tener entre 2 y 15 caracteres. Ingrese la marca nuevamente: ")

    precio=float(input("Ingrese el precio del vehículo: "))
    while precio <= 5000000:
        precio=float(input("El precio debe ser mayor a $5.000.000. Ingrese el precio nuevamente: "))

    multas={"monto": float(input("Ingrese el monto de las multas: ")),
              "fecha": input("Ingrese la fecha de las multas: ")}

    fecha_registro=input("Ingrese la fecha de registro del vehículo: ")
    dueño=input("Ingrese el nombre del dueño: ")
    region=input("Ingrese la región del vehículo (Ejemplo: Región Metropolitana): ")

    auto={"tipo": tipo, "patente": patente, "marca": marca, "precio": precio,
            "multas": multas, "fecha_registro": fecha_registro, "dueño": dueño, "region": region}

    autos.append(auto)
    print("Vehículo registrado correctamente.")

def buscar():
    patente_buscar=input("Ingrese la patente del vehículo a buscar: ")
    encontrado=False

    for auto in autos:
        if auto["patente"] == patente_buscar:
            print("\nInformación del vehículo:")
            print(f"Tipo: {auto['tipo']}")
            print(f"Patente: {auto['patente']}")
            print(f"Marca: {auto['marca']}")
            print(f"Precio: ${auto['precio']}")
            print(f"Multas: ${auto['multas']['monto']} - Fecha: {auto['multas']['fecha']}")
            print(f"Fecha de registro: {auto['fecha_registro']}")
            print(f"Dueño: {auto['dueño']}")
            print(f"Región: {auto['region']}")
            encontrado=True
            break

    if not encontrado:
        print("Vehículo no encontrado.")

def imprimir_certificados():
    for auto in autos:
        certificado_emision=round((1500 + 2000 * (auto['precio'] / 5000000)), 2)
        certificado_anotaciones=round(1500 + (2000 * (auto['precio'] / 5000000)), 2)
        certificado_multas=round(1500 + (2000 * (auto['precio'] / 5000000)), 2)

        print("\nCertificado de Emisión de Contaminantes:")
        print(f"Nombre del certificado: Emisión de contaminantes")
        print(f"Patente del auto: {auto['patente']}")
        print(f"Datos del dueño: {auto['dueño']}")
        print(f"Región: {auto['region']}")

        print("\nCertificado de Anotaciones Vigentes:")
        print(f"Nombre del certificado: Anotaciones vigentes")
        print(f"Patente del auto: {auto['patente']}")
        print(f"Datos del dueño: {auto['dueño']}")
        print(f"Región: {auto['region']}")

        print("\nCertificado de Multas:")
        print(f"Nombre del certificado: Multas")
        print(f"Patente del auto: {auto['patente']}")
        print(f"Datos del dueño: {auto['dueño']}")
        print(f"Región: {auto['region']}")
        print(f"Multas: ${auto['multas']['monto']} - Fecha: {auto['multas']['fecha']}")
        print(f"Emisión de Contaminantes: ${certificado_emision}")
        print(f"Anotaciones Vigentes: ${certificado_anotaciones}")

def main():
    while True:
        print("\n--- Menú ---")
        print("1. Grabar vehículo")
        print("2. Buscar vehículo")
        print("3. Imprimir certificados")
        print("4. Salir")

        opcion=input("Seleccione una opción (1-4): ")

        if opcion == "1":
            grabar()
        elif opcion == "2":
            buscar()
        elif opcion == "3":
            imprimir_certificados()
        elif opcion == "4":
            print("¡Gracias por utilizar el programa!")
            print("Automotora Auto Seguro - Camila Rebolledo")
            break
        else:
            print("Opción no válida. Intente nuevamente.")

if __name__ == "__main__":
    main()
```
